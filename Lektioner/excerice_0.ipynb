```python
import seaborn as sns

load_data = sns.load_dataset('mpg')
print(load_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print(load_data.describe())

sns.histplot(data=load_data, x='mpg', kde=True)
plt.title('Distribution of MPG')
plt.show()

sns.scatterplot(data=load_data, x='weight', y='mpg')
plt.title('Relationship between Weight and MPG')
plt.show()

sns.boxplot(data=load_data, x='origin', y='mpg')
plt.title('MPG by Origin')
plt.show()

sns.heatmap(load_data, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
